# Let's go making Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

#  Making Core Item

```python
def prediksi_harga_komoditas(target_city, target_commodity, dataset, window_size=7, epochs=50):
    """
    Fungsi End-to-End untuk memprediksi harga berdasarkan Kota & Komoditas.
    """
    print(f"\n{'='*50}")
    print(f"MEMULAI PROSES: {target_commodity} di {target_city}")
    print(f"{'='*50}")

    # --- 1. FILTERING DATA ---
    # Mengambil data hanya untuk kota dan komoditas yang dipilih
    df_filter = dataset[
        (dataset['city_name'] == target_city) &
        (dataset['commodity_name'] == target_commodity)
    ].copy()

    # Cek ketersediaan data
    if df_filter.empty:
        print(f"Error: Data tidak ditemukan untuk {target_commodity} di {target_city}.")
        return None

    # --- 2. PREPROCESSING ---
    # Urutkan waktu
    df_filter['date'] = pd.to_datetime(df_filter['date'])
    df_filter = df_filter.sort_values('date')

    # Pilih Fitur (Multivariate)
    # daily_price ditaruh di indeks 0 agar mudah diambil nanti
    fitur_cols = ['daily_price', 'avg_value', 'max_value', 'min_value', 'Month']
    df_ready = df_filter[fitur_cols].values

    # Normalisasi (Scaling) 0 s/d 1
    scaler = MinMaxScaler(feature_range=(0, 1))
    df_scaled = scaler.fit_transform(df_ready)

    # --- 3. SLIDING WINDOW ---
    X, y = [], []
    for i in range(window_size, len(df_scaled)):
        # X: Ambil data 7 hari ke belakang (semua fitur)
        X.append(df_scaled[i-window_size:i, :])
        # y: Ambil data hari ini (hanya daily_price / kolom ke-0)
        y.append(df_scaled[i, 0])

    X, y = np.array(X), np.array(y)

    # --- 4. DATA SPLITTING  ---
    split_idx = int(len(X) * 0.8)
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]

    print(f"Jumlah Data Train: {len(X_train)} | Data Test: {len(X_test)}")

    # --- 5. MEMBANGUN MODEL LSTM ---
    model = Sequential()
    # Layer LSTM 1
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(0.2))
    # Layer LSTM 2
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    # Output Layer (1 neuron = harga)
    model.add(Dense(units=1))

    model.compile(optimizer='adam', loss='mean_squared_error')

    # --- 6. TRAINING MODEL ---
    print("Sedang melatih model... (Tunggu sebentar)")
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_data=(X_test, y_test), verbose=0)
    print("Training Selesai!")

    # --- 7. EVALUASI & PREDIKSI ---
    predicted_prices = model.predict(X_test)

    # Mengembalikan skala asli (Inverse Transform)
    # Karena scaler kita multivariate (5 kolom), kita perlu trik untuk inverse
    def inverse_price(pred_array, scaler_obj):
        # Buat array kosong dengan bentuk yang sama seperti input scaler
        dummy_array = np.zeros((len(pred_array), 5))
        # Isi kolom pertama dengan hasil prediksi
        dummy_array[:, 0] = pred_array.flatten()
        # Inverse transform
        inv = scaler_obj.inverse_transform(dummy_array)
        # Ambil kembali kolom pertama saja (harga)
        return inv[:, 0]

    real_price = inverse_price(y_test.reshape(-1, 1), scaler)
    pred_price = inverse_price(predicted_prices, scaler)

    # Hitung Error
    rmse = np.sqrt(mean_squared_error(real_price, pred_price))
    mape = np.mean(np.abs((real_price - pred_price) / real_price)) * 100

    print(f"Performa Model -> RMSE: Rp {rmse:.2f} | MAPE: {mape:.2f}%")

    # --- 8. PREDIKSI BESOK (Forecast Next Day) ---
    # Ambil data window terakhir dari dataset asli
    last_window = df_scaled[-window_size:]
    last_window = last_window.reshape(1, window_size, 5) # Reshape ke (1, 7, 5)

    future_pred = model.predict(last_window)
    future_price_inv = inverse_price(future_pred, scaler)[0]

    print(f"Prediksi Harga Besok: Rp {future_price_inv:,.2f}")

    # --- 9. VISUALISASI ---
    plt.figure(figsize=(14, 6))

    # Ambil tanggal untuk sumbu X
    test_dates = df_filter['date'].iloc[split_idx + window_size:]

    plt.plot(test_dates, real_price, color='blue', label='Harga Asli (Aktual)')
    plt.plot(test_dates, pred_price, color='red', linestyle='--', label='Prediksi LSTM')

    plt.title(f"Prediksi Harga {target_commodity} di {target_city}\n(RMSE: {rmse:.2f})", fontsize=16)
    plt.xlabel('Tanggal')
    plt.ylabel('Harga (Rp)')
    plt.legend()
    plt.show()

    return model
```

```python
model_beras = prediksi_harga_komoditas(
    target_city="Jakarta Selatan",
    target_commodity="Beras IR. II (IR 64) Ramos",
    dataset=data,
    window_size=7,
    epochs=50
)
```
